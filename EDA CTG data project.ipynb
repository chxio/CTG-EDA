# Data Explanation
1. CTG data is Cardiotocogram data, Cardiotocogram is continuous recording of the fetal heart rate obtained via an ultrasound transducer placed on the mother's abdomen. CTG is widely used in pregnancy as a method of assessing fetal well‐being, predominantly in pregnancies with increased risk of complications.
2. CTG data include Accelerations, Deccelerations, Fetal Heart Rate over a period of time, Uterus contractions
3. The base variable which defines fetal state are - Accelerations (AC column), Deccelerations over 10 minute time period (DP, DR, DL, DS columns), Baseline Fetal Heart Rate (LB and LBE column), Uterus contractions (UC column), Heart rate variability in short term and long term (ASTV, MSTV, ALTV, MLTV columns), number of Uterus contractions in 10 minutes period.
4. After comparing with Satandard values, obstetrician determine fetal state (NSP column in data)

# T test

Hypothesis - population mean AC is near to zero for Unhealthy (State 2 or 3) Fetal state, we are trying to confirm that AC mean value is consistent over 2 different samples fopr Unhealthy fetal states

Filter out the dataset and divided the filtered data on 2 same size samples - data_group1, data_group2

Assumption - data_group1 and data_group2 has equal variances

P Value shows likelihood of finding a mean difference.

if calculated p value<0.05 then we reject the hypothesis

else we accept the hypothesis

```python
import scipy.stats as stats

df1 = df[df['NSP']>1]
data = df1["AC"]
data_group1 = data.sample(200)
data_group2 = data.sample(200)
 
# Perform the two sample t-test with equal variances
stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
```

Since P_value is greater than 0.05, we are accepting the hypothesis - AC mean zero is consistent over the population

We do not have sufficient evidence to say that the AC mean of Unhealthy fetal state between the two data groups is different.
